```python
import json
import pyiwn
from pyiwn import IndoWordNet
from heritage import HeritagePlatform
import stanza

iwn = IndoWordNet(lang=pyiwn.Language.SANSKRIT)
platform = HeritagePlatform()

try:
    stanza.download('sa')
    nlp = stanza.Pipeline('sa', processors='tokenize,pos,lemma,depparse')
except:
    nlp = None
```

```python

def get_heritage_analysis(word):
    try:
        analysis = platform.get_analysis(word, sentence=False, meta=True)
        readable_info = []
        for entry in analysis.values():
            if 'words' in entry:
                for word_list in entry['words']:
                    for word_data in word_list:
                        readable_info.append({
                            'Text': word_data.get('text', ''),
                            'Root': word_data.get('root', ''),
                            'Category': word_data.get('category', [])
                        })
        filtered_roots = [
            info['Root'] for info in readable_info
            if info['Root'] and not any(char.isdigit() for char in str(info['Root']))
        ]
        return readable_info, filtered_roots
    except Exception as e:
        return [], []
```

```python


def get_synsets_data(word):
    try:
        all_synsets = iwn.synsets(word)
        synsets_data = []
        for syn in all_synsets:
            try:
                syn_data = {
                    'synset': str(syn),
                    'examples': syn.examples(),
                    'gloss': syn.gloss(),
                    'head_word': syn.head_word(),
                    'lemma_names': syn.lemma_names(),
                    'lemmas': [str(lemma) for lemma in syn.lemmas()],
                    'pos': str(syn.pos())
                }
                synsets_data.append(syn_data)
            except Exception:
                continue
        return synsets_data
    except Exception:
        return []
```

```python

def process_sentence(sentence_text):
    sentence_data = {
        'original_sentence': sentence_text,
        'annotation': []
    }

    if nlp:
        try:
            doc = nlp(sentence_text)
            for sentence in doc.sentences:
                for word in sentence.words:
                    heritage_info, filtered_roots = get_heritage_analysis(word.text)
                    word_analysis_data = []

                    for info in heritage_info:
                        root_synsets = get_synsets_data(info['Root'])
                        word_analysis_data.append({
                            'root': info['Root'],
                            'category': info['Category'],
                            'synset_data': root_synsets
                        })

                    annotation_entry = {
                        'text': word.text,
                        'deprel': word.deprel,
                        'pos': word.pos,
                        'upos': word.upos,
                        'feats': word.feats,
                        'word_analysis': word_analysis_data
                    }

                    sentence_data['annotation'].append(annotation_entry)
        except Exception:
            pass

    return sentence_data
```

```python
def main():
    all_data = {
        'sentences': []
    }

    try:
        with open('data.txt', 'r', encoding='utf-8') as f:
            sentences = f.readlines()

        sentences = [sentence.strip() for sentence in sentences if sentence.strip()]

        if not sentences:
            print("data.txt is empty")
            return

        print(f"Found {len(sentences)} sentences in data.txt")

        for i, sentence in enumerate(sentences):
            print(f"Processing sentence {i+1}: {sentence}")
            try:
                sentence_result = process_sentence(sentence)
                all_data['sentences'].append(sentence_result)
                print(f"Successfully processed sentence {i+1}")
            except Exception as e:
                print(f"Error processing sentence {i+1}: {e}")
                all_data['sentences'].append({
                    'original_sentence': sentence,
                    'error': str(e)
                })

        output_filename = 'sanskrit.json'
        with open(output_filename, 'w', encoding='utf-8') as f:
            json.dump(all_data, f, ensure_ascii=False, indent=2)
        print(f"Data saved to {output_filename}")

    except FileNotFoundError:
        print("data.txt file not found in current directory")

    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```
